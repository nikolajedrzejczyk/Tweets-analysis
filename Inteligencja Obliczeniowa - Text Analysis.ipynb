```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from datetime import timedelta, date
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

CREATING A FUNCTION THAT LISTS SUCCESSIVE DATES FROM WHICH TWEETS WILL BE DOWNLOADED

```python
def all_dates(start_date, end_date):
    for n in range(int((end_date - start_date).days)): #I create loops with the length of the number of days from which data is downloaded
        yield start_date + timedelta(n) #yield - lists the days one by one, adding one day at a time
```

DEFINING START AND END DATES, MAXIMUM NUMBER OF TWEETS DOWNLOADED PER DAY AND CREATING EMPTY LISTS TO SAVE DOWNLOADED TWEETS

```python
#I use the downloaded library to define dates - in order: year, month, day
start_date = date(2023, 2, 10)
end_date = date(2023, 3, 20)

#I set a maximum of 1000 tweets per day
max_tweets = 1000

#I create empty lists needed in the next part
tweets_list1 = []
tweets_list2 = []
tweets_list3 = []
```

DOWNLOADING TWEETS CONTAINING "Selena Gomez"

```python
#For each subsequent date: I define it as start and end: this date + 1 day

for date in all_dates(start_date, end_date):
    start = date.strftime("%Y-%m-%d")
    end = (date + timedelta(days = 1)).strftime("%Y-%m-%d")
    query1 = f'Selena Gomez since:{start} until:{end} lang:en' #lang:en - because I need only english tweets

    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(query1).get_items()):
        if i + 1 >= max_tweets: #The condition is that after downloading 1000 tweets it will go to the next day
            break
        tweets_list1.append([tweet.date, tweet.content]) #I save the date and content of each tweet to the lists I created earlier
```

CREATE A DATAFRAME FROM DOWNLOADED TWEETS AND SAVE TO A FILE

```python
#Creation DataFrame
tweets_df1 = pd.DataFrame(tweets_list1, columns = ['Datetime', 'Text'])
tweets_df1.to_csv(f'SelenaGomez.csv', sep = ',', index = False)
tweets_df1
```

# Preprocessing:

NORMALIZATION:

```python
import re

def remove_sign(tweet):
    processed_tweet = re.sub(r'@\w+', '', tweet) #removing mentions
    processed_tweet = re.sub(r'#', '', processed_tweet) #removing hashtags
    processed_tweet = re.sub(r'https?://\S+', '', processed_tweet) #removing URL
    processed_tweet = re.sub(r"'\w+", '', processed_tweet) #removing endings: 's, 'm, etc.
    return processed_tweet

def normalized(tweet):
    normalized_text = tweet.lower() #change all letters to lowercase
    return normalized_text
```

```python
df_Selena['Text'] = df_Selena['Text'].apply(remove_sign)
df_Selena
```

```python
df_Hailey['Text'] = df_Hailey['Text'].apply(remove_sign)
df_SelenaAndHailey['Text'] = df_SelenaAndHailey['Text'].apply(remove_sign)
```

```python
df_Selena['Text'] = df_Selena['Text'].apply(normalized)
df_Selena
```

```python
df_Hailey['Text'] = df_Hailey['Text'].apply(normalized)
df_SelenaAndHailey['Text'] = df_SelenaAndHailey['Text'].apply(normalized)
```

TOKENIZATION AND REMOVAL STOP WORDS:

```python
from nltk.corpus import stopwords

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))  #language selection to English
    
    word_tokens = nltk.word_tokenize(text)

    filtered_words = [w for w in word_tokens if not w in stop_words] #removing stop words for lemmatization
    cleaned_words = [w for w in filtered_words if w.isalpha()] #removing commas, periods, etc.

    return cleaned_words
```

```python
df_Selena['Text'] = df_Selena['Text'].apply(remove_stopwords)
df_Selena
```

```python
df_Hailey['Text'] = df_Hailey['Text'].apply(remove_stopwords)
df_SelenaAndHailey['Text'] = df_SelenaAndHailey['Text'].apply(remove_stopwords)
```

LEMMATIZATION:

```python
lemmatizer = nltk.WordNetLemmatizer()
def lemmatize_text(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]
```

```python
df_Selena['Text'] = df_Selena['Text'].apply(lemmatize_text)
df_Selena
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

#Sentiment analyzer initialization
sid = SentimentIntensityAnalyzer()

#A function for analyzing sentiment and returning a label
def get_sentiment_label(text):
    scores = sid.polarity_scores(text)
    compound_score = scores['compound']
    
    if compound_score >= 0.05:
        return "Positive"
    elif compound_score <= -0.05:
        return "Negative"
    else:
        return "Neutral"
    
def get_score(text):
    scores = sid.polarity_scores(text)
    compound_score = scores['compound']
    return compound_score
```

```python
df_Selena['sentiment_label'] = df_Selena['Text'].apply(get_sentiment_label)
df_Selena['score'] = df_Selena['Text'].apply(get_score)
df_Selena
```

# Second way - NRC Lexicon

```python
from nrclex import NRCLex
import plotly.express as px

#Emotion classification function
def classify_emotions(text):
    emotions = NRCLex(text).affect_frequencies  #I delete categories that I am not interested in
    emotions.pop("positive", None)
    emotions.pop("negative", None)
    emotions.pop("anticip", None)
    emotions.pop("anticipation", None)
    return emotions
```
